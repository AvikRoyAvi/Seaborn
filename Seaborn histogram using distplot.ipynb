```python
import seaborn as sns
from scipy.stats import norm
from matplotlib import pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.distplot(tips_df['tip'])
plt.show()
```

```python
sns.histplot(tips_df['size'])
plt.show()
```

```python
sns.histplot(tips_df['total_bill'])
plt.show()
```

```python
sns.histplot(tips_df['total_bill'], bins = 50)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], hist = False)
plt.show()
```

```python
sns.histplot(tips_df['total_bill'], kde = False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], fit = norm, kde = False)
plt.show()
```

```python
sns.histplot(tips_df['total_bill'], color = 'r', kde = True)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], color = 'r', kde = True, vertical = True)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], color = 'r', kde = True, axlabel='Total Bill')
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.set()
sns.histplot(tips_df['total_bill'], label='Total Bill', kde = True)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize = (16, 9))
sns.set()
sns.histplot(tips_df['total_bill'], bins = bins, kde = True)
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.set()
sns.distplot(tips_df['total_bill'], kde = True,
             hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5, 'linestyle':'--', 'alpha':0.9})
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.set()
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df['total_bill'], kde = True,
             hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5, 'linestyle':'--', 'alpha':0.9}, 
             kde_kws = {'color':'#8e00ce', 'linewidth':8, 'linestyle':'--', 'alpha':0.9},)
plt.xticks(bins)
plt.title("Histogram of Restorant Total Bill", fontsize = 20)
plt.xlabel("Total Bill", fontsize = 15)
plt.legend()
plt.show()
```

```python
# Plot multiple seaborn histogram in single graph
plt.figure(figsize = (16, 9))
sns.histplot(tips_df["total_bill"], bins = 9, label = "total_bil", kde = True)
sns.histplot(tips_df["tip"], bins = 9, label = "tip", kde = True)
sns.histplot(tips_df["size"], bins = 9, label = "size", kde = True)
 
plt.legend()
plt.show()
```
