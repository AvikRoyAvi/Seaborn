```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df
```

```python
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df)
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, hue_order = ['female', 'male'])
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, style = 'who')
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, style = 'who', size = 'who')
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, style = 'who', size = 'who', sizes = (100, 400))
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, style = 'who', size = 'who', sizes = (100, 400))
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, style = 'who', size = 'who', sizes = (100, 400), palette= 'hot')
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, style = 'who', size = 'who', sizes = (100, 400), markers = ['3', '1', 3])
plt.show()
```

```python
ax = sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, style = 'who', size = 'who', sizes = (100, 400), palette= 'hot')
ax.set(xlabel = "Age",
      ylabel = "Fare",
      title = "Seaborn Scatter Plot of Age and Fare")
plt.show()
```

```python
plt.figure(figsize = (16, 9))
kwargs = {'edgecolor': "r", 'facecolor': "k", 'linewidth': 2.7, 'linestyle': '--',}
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df, style = 'who', size = 'who', **kwargs)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'who', y = 'fare', hue = 'alive', data = titanic_df, style = 'alive', size = 'who', sizes = (100, 400))
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex", palette = "magma", size = "who", sizes = (50, 300))
plt.title("Scatter Plot of Age and Fare", fontsize = 25)
plt.xlabel("Age", fontsize = 20)
plt.ylabel("Fare", fontsize = 20)
plt.savefig("Scatter Plot of Age and Fare")
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
sns.lineplot(x = 'age', y = 'fare', data = titanic_df)
sns.barplot(x = 'age', y = 'fare', data = titanic_df)
plt.savefig("Scatter Plot, Line Plot, Bar Plot in one Plot.")
plt.show()
```
