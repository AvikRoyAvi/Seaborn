```python
globalWarming_df = pd.read_csv('D:\Python codes\Seaborn\Who_is_responsible_for_global_warming.csv')
globalWarming_df
```

```python
globalWarming_df.corr()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df.corr(), annot = True)
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(globalWarming_df.corr(), annot = True, linewidths = 3)
ax.tick_params(size = 10, color = 'w', labelsize = 10, labelcolor = 'w')
plt.title('Heat map of "Who is responsible for global warming"', fontsize = 25)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
corr_mx = globalWarming_df.corr()
matrix = np.tril(corr_mx)
sns.heatmap(corr_mx, mask=matrix)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
corr_mx = globalWarming_df.corr()
matrix = np.triu(corr_mx)
sns.heatmap(corr_mx, mask=matrix)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                         columns = np.append(cancer_dataset['feature_names'], ['target']))
cancer_df
```

```python
plt.figure(figsize = (25, 25))
ax = sns.heatmap(cancer_df.corr(), annot = True, linewidths = 3)
ax.tick_params(size = 10, color = 'w', labelsize = 10, labelcolor = 'w')
plt.title('Heat map of "Breast Cancer Patients"', fontsize = 25)
plt.show()
```
