```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.barplot()
plt.show()
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
plt.show()
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)
plt.show()
```

```python
# Pass dataset using data parameter
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df)
plt.show()
```

```python
order = ['Sun', 'Thur', 'Fri', 'Sat']
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, order = order)
plt.show()
```

```python
hue_order = ['Female', 'Male']
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, hue_order = hue_order)
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, estimator = np.min)
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, estimator = np.max)
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, ci = 20)
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, n_boot = 2)
plt.show()
```

```python
sns.barplot(y = 'day', x = 'total_bill', hue = 'sex', data = tips_df)
plt.show()
```

```python
sns.barplot(x = 'total_bill', y = 'size', hue = 'sex', data = tips_df, orient = 'h')
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, color = 'r')
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, color = 'g')
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, palette = 'hot')
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, palette = 'magma')
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, palette = 'magma', saturation = .3)
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, palette = 'hot', errcolor = '0.8', errwidth = '10')
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, dodge = False)
plt.show()
```

```python
kwargs = {'alpha': 0.9, 'linestyle': ':', 'linewidth': 5, 'edgecolor': 'k'}
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, **kwargs)
plt.show()
```

```python
sns.set()
kwargs = {'alpha': 0.9, 'linestyle': ':', 'linewidth': 5, 'edgecolor': 'k'}
sns.barplot(x = 'day', y = 'total_bill', data = tips_df, **kwargs)
plt.show()
```

```python
sns.set()
sns.barplot(x = 'day', y = 'total_bill', data = tips_df, alpha = 0.6, linestyle = '--', linewidth = 8, edgecolor = 'w')
plt.show()
```

```python
sns.set()
ax = sns.barplot(x = 'day', y = 'total_bill', data = tips_df, alpha = 1, linestyle = '-.', linewidth = 3, edgecolor = 'g')
ax.set(title = "Barplot of Tips DataFrame",
      xlabel = "Days",
      ylabel = "Total Bill")
plt.show()
```

```python
sns.set()
plt.figure(figsize = (16, 9))
sns.barplot(x = 'day', y = 'total_bill', data = tips_df, alpha = 0.6, linestyle = '--', linewidth = 8, edgecolor = 'k')
plt.title("Barplot of Days and Total Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
plt.savefig("Barplot of Days and Total Bill")
plt.show()
```
